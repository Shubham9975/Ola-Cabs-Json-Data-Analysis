```python
import pandas as pd
import json
from textblob import TextBlob
import matplotlib.pyplot as plt
```

<h2>Reading JSON file</h2>

```python
with open('Ola (1).json', 'r', encoding='utf-8') as file:
    data = json.load(file)
```

<h2>Flattenning Tweet for parsing inner json tags</h2>

<h2>Also Adding new metrics(Columns) for further Analysis</h2>

```python
def flatten_tweet(tweet):
    return {
        "tweet_id": tweet["tweet_id"],
        "content": tweet["content"],
        "datetime": tweet["datetime"],
        "likes": tweet["likes"],
        "shares": tweet["shares"],
        "views": tweet["views"],
        "source": tweet["source"],
        "isBlue": tweet["isBlue"],
        "followers": tweet["followers"],
        "hashtags": ', '.join(tweet["hashtags"]) if tweet["hashtags"] else None,
        "location": tweet["location"],
        "username": tweet["username"],
        "url": tweet["url"],
        "user_mentions_names": ', '.join([mention["name"] for mention in tweet["user_mentions"]]) if tweet["user_mentions"] else None,
        "user_mentions_screen_names": ', '.join([mention["screen_name"] for mention in tweet["user_mentions"]]) if tweet["user_mentions"] else None,
        "media_urls": ', '.join(tweet["media"]) if tweet["media"] else None,
        "comments_count": len(tweet["comments"]),
        "comments_content": '; '.join([comment["content"] for comment in tweet["comments"]]) if tweet["comments"] else None
    }

tweets_df = pd.DataFrame([flatten_tweet(tweet) for tweet in data])
tweets_df.head()
```

<h2>Storing Raw Data to CSV file</h2>

```python
tweets_df.to_csv("parsed_json_to_csv.csv", index=False)
```

```python
tweets_df.info()
```

<h2>Drop Unnecessary Columns</h2>

```python
columns_to_drop = ["isBlue", "location", "user_mentions_screen_names", "comments_content", "url"]
tweets_df = tweets_df.drop(columns=columns_to_drop)
```

```python
tweets_df.info()
```

<h2>Correcting Data Types of columns which are wrongly parsed</h2>

```python
tweets_df['datetime'] = pd.to_datetime(tweets_df['datetime'], format='%d-%m-%Y %H:%M:%S')

tweets_df['views'] = tweets_df['views'].fillna(0)

tweets_df['views'] = tweets_df['views'].astype('int64')
```

<h2>Keeping relevant Tweets for better analysis</h2>

```python
keywords = ['ola cabs', 'olacabs', 'ola support', 'cabs']
keywords_mentions = ['ola cabs', 'olacabs', 'ola support']

content_filter = tweets_df['content'].str.contains('|'.join(keywords), case=False, na=False)

mentions_filter = tweets_df['user_mentions_names'].str.contains('|'.join(keywords_mentions), case=False, na=False)

relevant_tweets_df = tweets_df[content_filter | mentions_filter]

print(f"Total relevant tweets: {len(relevant_tweets_df)}")
```

<h2>Removing irrelevant Tweets for better analysis</h2>

```python
keywords = ["ola electric", "ola foods", "olaelectric", "olafoods", 'Olamide', 'ola Scooter', 'olaScooter', 'OLABuilderState', 'layoffs']

filtered_df = relevant_tweets_df[
    ~(
        relevant_tweets_df['content'].str.contains('|'.join(keywords), case=False, na=False) |
        relevant_tweets_df['user_mentions_names'].str.contains('|'.join(keywords), case=False, na=False)
    )
]

filtered_df.info()
```

<h2>Storing cleaned dataset into CSV</h2>

```python
filtered_df.to_csv("cleaned_tweets.csv", index=False)
```

<h2>Sentiment Analysis using TextBlob library and categorized into Positive, Negative or Neutral and calculates percentage of each category</h2>

```python
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity == 0:
        return "Neutral"
    else:
        return "Negative"

filtered_df['sentiment'] = filtered_df['content'].apply(get_sentiment)

sentiment_counts = filtered_df['sentiment'].value_counts(normalize=True) * 100
sentiment_counts
```

<h2>Extracting posts with issues like drivers, cancellations and customer service</h2>

```python
filtered_df = filtered_df.copy()
complaints_keywords = ["driver", "cancel", "customer service", "issue"]
filtered_df['is_complaint'] = filtered_df['content'].apply(lambda x: any(keyword in x.lower() for keyword in complaints_keywords))

complaints = filtered_df[filtered_df['is_complaint']]
```

<h2>Categorizing based on type of complaint</h2>

```python
def categorize_complaint(text):
    if "driver" in text.lower():
        return "Driver Issues"
    elif "cancel" in text.lower():
        return "Booking Cancellations"
    elif "customer service" in text.lower() or "issue" in text.lower():
        return "Customer Service"
    else:
        return "Other"
        
complaints = complaints.copy()
complaints['complaint_category'] = complaints['content'].apply(categorize_complaint)
```
