```python
import json
import pandas as pd
```

```python
with open("Ola (1).json", "r", encoding="utf-8") as f:
    data =  json.load(f)
```

```python
df = pd.json_normalize(data)
```

```python
df.head()
```

```python
df['hashtags'] = df['hashtags'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
df['media'] = df['media'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)
```

```python
df.head()
```

```python
df_exploded = df.explode('user_mentions')
```

```python
df_exploded.head()
```

```python
mentions_df = pd.json_normalize(df_exploded['user_mentions'])
```

```python
mentions_df.head()
```

```python
mentions_df['tweet_id'] = df_exploded['tweet_id'].values
```

```python
mentions_df.head()
```

```python
df_exploded = df.explode('comments')
```

```python
df_exploded.head()
```

```python
comments_df = pd.json_normalize(df_exploded['comments'])
```

```python
comments_df['tweet_id'] = df_exploded['tweet_id'].values
```

```python
comments_df.head()
```

```python
mentions_df = mentions_df[['tweet_id', 'name', 'screen_name']]
```

```python
mentions_df.head()
```

```python
mentions_df.rename(columns={'name': 'mention_name', 'screen_name': 'mention_screen_name'}, inplace=True)
```

```python
mentions_df.head()
```

```python
comments_df.head()
```

```python
comments_df = comments_df[['tweet_id', 'content']]
comments_df.rename(columns={'content': 'comments'}, inplace=True)
```
